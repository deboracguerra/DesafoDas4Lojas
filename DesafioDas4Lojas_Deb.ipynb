#1. Análise do faturamento



```python
import pandas as pd

df = loja
print(df['Preço'].sum())

df = loja2
print(df['Preço'].sum())

df = loja3
print(df['Preço'].sum())

df = loja4
print(df['Preço'].sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

df_faturamento = pd.DataFrame({
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Faturamento': [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
})

plt.figure(figsize=(10, 6))
sns.barplot(x='Faturamento', y='Loja', data=df_faturamento, palette='viridis')
plt.title('Faturamento Total por Loja (Barras Horizontais)')
plt.xlabel('Faturamento Total')
plt.ylabel('Loja')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

# 2. Vendas por Categoria

```python
def contar_por_categoria(df):
    return df['Categoria do Produto'].value_counts()

print('--- Vendas por Categoria da Loja 1 ---')
top_loja1 = contar_por_categoria(loja).head(10)
print(top_loja1)

print('--- Vendas por Categoria da Loja 2 ---')
top_loja2 = contar_por_categoria(loja2).head(10)
print(top_loja2)

print('--- Vendas por Categoria da Loja 3 ---')
top_loja3 = contar_por_categoria(loja3).head(10)
print(top_loja3)

print('--- Vendas por Categoria da Loja 4 ---')
top_loja4 = contar_por_categoria(loja4).head(10)
print(top_loja4)
```

## Recomendação

### Vender: **Loja 4**

**Justificativas principais (baseadas nas análises e gráficos):**

1. **Menor faturamento absoluto** (Gráfico de barras de receita): limita o potencial de contribuição ao caixa da rede.
2. **Vantagem logística isolada** (frete baixo), **sem reflexo** proporcional em vendas (Gráfico de dispersão frete × receita): eficiência que **não vira demanda**.
3. **Mix com sinal de desalinhamento** (Gráfico de Pareto dos menos vendidos): **muitos itens caros** (geladeira, lavadora, guarda-roupas, guitarra/violão) entre os piores, sugerindo **estoque arriscado** e **capital imobilizado**.

**Contrapontos considerados:**

* O frete mais barato **é força real**; porém, **não foi suficiente** para alavancar o faturamento.
* Há alguns best-sellers (cama box, cama king), mas o **portfólio é heterogêneo** e o **posicionamento pouco claro** frente às outras lojas, que têm identidades mais nítidas (L3 “coração de móveis”; L1 “móveis + linha branca + TV” de alto giro; L2 “diversificada, boa reputação”).
